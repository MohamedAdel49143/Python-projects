## 17. Replace Salary Values With 0 and 1


```python
adult.columns
```

```python
adult['income'].value_counts()
```

```python
sns.countplot(data=adult,x='income')
```

```python
def salary_data(sal):
    if sal=='<=50K':
        return 0
    else:
        return 1
```

```python
#adult['encoded_salary'] = adult['income'].apply(salary_data)
adult['income'].apply(salary_data)
```

```python
adult.head(1)
```

```python
#another answer
adult['income'].replace(['<=50K','>50K'],[0,1],inplace=True)
```
